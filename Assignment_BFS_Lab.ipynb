```python
from collections import deque

def bfs_path(graph, start, goal):
    """
    Finds the shortest path from start to goal using Breadth-First Search.
    Explores neighbors in alphabetical order.
    """
    # The queue stores tuples of (node, path_to_node)
    # We use deque for an efficient pop from the left (popleft)
    queue = deque([(start, [start])])

    # Keep track of visited nodes.
    # We add to 'visited' when we add to the queue to prevent
    # adding the same node multiple times.
    visited = {start}

    while queue:
        # Get the node and path from the front of the queue
        (current_node, path) = queue.popleft()

        # --- Goal Check ---
        if current_node == goal:
            return path  # We found the goal!

        # --- Add Neighbors to Queue ---
        # Get neighbors and sort them alphabetically
        neighbors = sorted(graph[current_node])

        for neighbor in neighbors:
            if neighbor not in visited:
                # Mark as visited *before* adding
                visited.add(neighbor)
                # Add the neighbor and the new path to the back of the queue
                queue.append((neighbor, path + [neighbor]))

    # If the queue is empty and we haven't found the goal
    return "No path found"

# --- Main Program ---

# Graph representation for the new image
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'C', 'D', 'E'],
    'C': ['A', 'B', 'E'],
    'D': ['B', 'E', 'F'],
    'E': ['C', 'B', 'D', 'F'],
    'F': ['E', 'D']
}

# Find the path
start_node = 'A'
goal_node = 'F'
path = bfs_path(graph, start_node, goal_node)

print(f"Breadth-First Search (BFS) Path:")
print(f"From: {start_node} To: {goal_node}")
if isinstance(path, list):
    print(f"Path: {' -> '.join(path)}")
else:
    print(path)
```
